```python
import pandas as pd
import numpy as np
```

```python
pd.__version__
```

```python
# import data from csv to pd df

url="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Confirmed.csv"
df_con=pd.read_csv(url)
url2="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Deaths.csv"
df_dth=pd.read_csv(url2)
url3="https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_19-covid-Recovered.csv"
df_rec=pd.read_csv(url3)
```

```python
# subset data from imported data

df_con2 = df_con.loc[df_con['Country/Region'] == "Australia"]
df_con2

df_dth2 = df_dth.loc[df_dth['Country/Region'] == "Australia"]
df_dth2

df_rec2 = df_rec.loc[df_rec['Country/Region'] == "Australia"]
df_rec2
```

```python
# unpivot data based on a multiple key

df_con3 = pd.melt(df_con2, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Confirmed_Cases')
df_con3

df_dth3 = pd.melt(df_dth2, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Death_Cases')
df_dth3

df_rec3 = pd.melt(df_rec2, id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], var_name='Date', value_name='Recovered_Cases')
df_rec3
```

```python
# integrate data based on a multiple key

integrated_df = pd.merge(df_con3, df_dth3,  how='left', left_on=['Province/State','Country/Region','Lat','Long','Date'], right_on = ['Province/State','Country/Region','Lat','Long','Date'])
integrated_df

integrated_df2 = pd.merge(integrated_df, df_rec3,  how='left', left_on=['Province/State','Country/Region','Lat','Long','Date'], right_on = ['Province/State','Country/Region','Lat','Long','Date'])
integrated_df2
```

```python
# cleaning

integrated_df3 = integrated_df2[['Date', 'Country/Region', 'Province/State', 'Lat', 'Long', 'Confirmed_Cases', 'Death_Cases', 'Recovered_Cases']]
integrated_df3

integrated_df3[['Month','Day','Year']] = integrated_df3.Date.str.split("/",expand=True,)
integrated_df3

integrated_df3['TimeSeries']=integrated_df3['Day'].astype(str)+'/'+integrated_df3['Month']+'/'+integrated_df3['Year']
integrated_df3

integrated_df3.drop(['Date'], axis=1)

integrated_df3 = integrated_df3[['TimeSeries', 'Country/Region', 'Province/State', 'Lat', 'Long', 'Confirmed_Cases', 'Death_Cases', 'Recovered_Cases', 'Day', 'Month', 'Year']]
integrated_df3
```
